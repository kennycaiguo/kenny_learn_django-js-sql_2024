## Python tuple[tʌpl]

### A tuple is a collection which is ordered and unchangeable.
### be careful,if a tuple only has a element,you have to write it so: (1,) . don't forget the ","

```python
t = (1,2,3)
# get tuple element: just like the list 
print(t[1])
```

```python
#Like List Traversal, we can traverse through a tuple using for loop.
for i in t:
    print(i)
```

```python
# To Concatenation of Python Tuples, we will use plus operators(+). (1,2,3)+(4,5,6)
print((1,2,3)+(3,4,5))
```

```python
# A nested tuple in Python means a tuple inside another tuple. 
# example:
# tup1 = (0, 1, 2, 3)
# tup2 = ('python', 'geek')

# tup3 = (tup1, tup2)

t1=("Kenny","Jade","Brendy")
t2=(100,80,70)
t3 = (t1,t2)
print(t3)




```

```python
# Repetition Python Tuples
# We can create a tuple of multiple same elements from a single element in that tuple. 
# (1,)*3  # (1, 1, 1)
print((3,)*5)
```

```python
# Slicing Tuples in Python ,just like the list
t = (1, 2, 3, 3, 4, 5)
print(t[1:4])
```

## Sets in Python

### A Set in Python is used to store a collection of items with the following properties.

### No duplicate elements. If try to insert the same item again, it overwrites previous one.
### An unordered collection. When we access all items, they are accessed without any specific order and we cannot access items using indexes as we do in lists.
### Internally use hashing that makes set efficient for search, insert and delete operations. It gives a major advantage over a list for problems with these operations.
### Mutable, meaning we can add or remove elements after their creation, the individual elements within the set cannot be changed directly.

```python
# create a set
# s = {1,2,3}
# print(s,type(s))
# s1=set([4,5,6])
# print(s1,type(s1))
```

```python
# Set is unordered
s = {"apple", "banana", "cherry"}
print(s)
```

```python
# set not allows duplicated elements
s = {"Hello","World","Hello","Geeks"}
print(s)
```

```python
# set iteration
# s = {"Hello","World","Hello","Geeks"}
for ss in s:
    print(ss)
```

```python
# Set methods,set don't support + operator
# {1,2}+{3,5} # don't support + operator
# add,to put an element into the set
# s1 = {1,2}
# s1.add(3)
# print(s1)
```

```python
# union() or | operator
s1 = {10,20}
s2 = {20,30}
# print(s1.union(s2))
print(s1 | s2)
```

```python
# intersection() or & operator
# s1={1,2,3,4}
# s2={3,4,5,6}
# print(s1.intersection(s2))
# print(s1 & s2)
```

```python
# difference() or - operator
# s1={1,2,3,4,5}
# s2={3,4,5,6,7}   
# print(s1.difference(s2))
print(s1-s2)
```

```python
# Clearing Python Sets
s={0,1,2,3}
s.clear()
print(s)
```

```python
# Python Frozen Sets
# Frozen sets in Python are immutable objects that only support methods and operators that produce a result without affecting the frozen set or sets to which they are applied. It can be done with frozenset() method in Python.

# While elements of a set can be modified at any time, elements of the frozen set remain the same after creation. 

# If no parameters are passed, it returns an empty frozenset.

# A frozen set
fs = frozenset(["e", "f", "g"])

# print("\nFrozen Set")
print(fs,type(fs))
# the following will get an error
# fs.add('h')
```

## Operators for Sets
### Sets and frozen sets support the following operators:

#### Operators	Notes
#### key in s	containment check
#### key not in s	non-containment check
#### s1 == s2	s1 is equivalent to s2
#### s1 != s2	s1 is not equivalent to s2
#### s1 <= s2	s1 is subset of s2
#### s1 < s2	s1 is proper subset of s2
#### s1 >= s2	s1 is superset of s2
#### s1 > s2	s1 is proper superset of s2
#### s1 | s2	the union of s1 and s2
#### s1 & s2	the intersection of s1 and s2
#### s1 - s2	the set of elements in s1 but not s2
#### s1 ˆ s2	the set of elements in precisely one of s1 or s2

## Python dictionary
###    Python dictionary is a data structure that stores the value in key: value pairs. Values in a dictionary can be of any data type and can be duplicated, whereas keys can't be repeated and must be immutable.



```python
# create a dict
dic1 = {"name":'Jackline',"age":20}
print(dic1,type(dic1))
dic2 = dict(name="Jade",gender="female",age=25)
print(dic2,type(dic2))
```

```python
# get dict item: 
# 1>dict[key]
dic1 = {"name":'Jackline',"age":20}
print(dic1["age"])   
# 2> dict.get(key)
dic2 = dict(name="Jade",gender="female",age=25)
print(dic2.get("gender"))
```

```python
# Adding and Updating Dictionary Items
# We can add new key-value pairs or update existing keys by using assignment.
# if you use a key that is not currently exist in the dict,you will create a new item  with the key and the value you provided
dic1["gender"]="male"
print(dic1)
# if you use a key that is  currently exist in the dict,you will update the item that has the key with the value you provided
dic1["gender"]="female"
print(dic1)
```

```python
# Removing Dictionary Items
# We can remove items from dictionary using the following methods:

# del: Removes an item by key.
# pop(): Removes an item by key and returns its value.
# clear(): Empties the dictionary.
# popitem(): Removes and returns the last key-value pair.

# example:
d = {1: 'Geeks', 2: 'For', 3: 'Geeks', 'age':22}

# # Using del to remove an item
# del d["age"]
# print(d)

# # Using pop() to remove an item and return the value
# val = d.pop(1)
# print(val)

# # Using popitem to removes and returns
# # the last key-value pair.
key, val = d.popitem()
print(f"Key: {key}, Value: {val}")

# # Clear all items from the dictionary
d.clear()
print(d)
```
